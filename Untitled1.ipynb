````python

import openai
import pandas as pd
import time
import json
from openai import AzureOpenAI 

AZURE_OPENAI_KEY = ""
AZURE_OPENAI_ENDPOINT = ""
AZURE_DEPLOYMENT_NAME = ""  # e.g., gpt-4o-mini
AZURE_API_VERSION = ""  # or whatever you're using

# Initialize Azure OpenAI Service client with key-based authentication    
client = AzureOpenAI(  
    azure_endpoint=AZURE_OPENAI_ENDPOINT,  
    api_key=AZURE_OPENAI_KEY,  
    api_version=AZURE_API_VERSION,
)

# Your DataFrame
df = pd.read_csv("output.csv")

def prompt_for_narration(narration, direction):
    return f"""
You are a financial data extractor. Given a bank transaction narration, extract the following:

- sender_or_reciever_info: who is the sender or receiver (or UPI ID if available)
- transaction_type: method of transaction like NEFT, UPI, IMPS, etc.
- currency: if foreign currency involved, name it (e.g. USD, EUR)
- exchange_rate: if mentioned, give exchange rate as number

The transaction direction is '{direction}' (either "credit" or "debit").

Narration:
\"{narration}\"

please give all 4 details for statement if not present, give NaN
Example:
given bank statement: 
NEFT CR-IBKO0000001-REANZEN ENTERPRISE INDIA PRIVATE LIMITED-ZENEDL CNS PVT LTD-07240403NEFT0128
output:
sender_or_reciever_info:REANZEN ENTERPRISE INDIA PRIVATE LIMITED
transaction_type: NEFT,
currency: NaN,
exchange_rate: NaN,

"""


def call_gpt(narration, direction):
    prompt = prompt_for_narration(narration, direction)
    try:
        response = client.chat.completions.create(
            model=AZURE_DEPLOYMENT_NAME,
            messages=[{"role": "user", "content": prompt}],
            temperature=0.2,
            max_tokens=200,
            top_p=1.0,
            frequency_penalty=0.0,
            presence_penalty=0.0,
            stop=["\n"]
        )

        # Print the full response for debugging
        print("Response from OpenAI:", response)

        # Now, response is a `ChatCompletion` object, so we need to access the text correctly
        content = response.choices[0].message.content

        if not content or content == "```json":
            print(f"Empty or invalid response for narration: {narration[:30]}...")
            return {
                "sender_or_reciever_info": None,
                "transaction_type": None,
                "currency": None,
                "exchange_rate": None
            }

        print("Content from response:", content)

        try:
            # Try parsing the response as JSON
            extracted_features = json.loads(content)
            return extracted_features
        except json.JSONDecodeError:
            # If the response isn't valid JSON, return None for all features
            print(f"Error parsing response: {content}")
            return {
                "sender_or_reciever_info": None,
                "transaction_type": None,
                "currency": None,
                "exchange_rate": None
            }
    except Exception as e:
        print(f"Error processing narration: {narration[:30]}... -> {e}")
        return {
            "sender_or_reciever_info": None,
            "transaction_type": None,
            "currency": None,
            "exchange_rate": None
        }

# Apply row-by-row with progress
results = []
for i, row in df.iterrows():
    narration = row.get('TransactionRemarks') or row.get('Narration') or row.get('Details') or ''
    
    # Determine direction
    credit = row.get('Deposit(Cr)', '') or row.get('Credit', '')
    debit = row.get('Withdrawl(Dr)', '') or row.get('Debit', '')
    direction = 'credit' if pd.notnull(credit) and str(credit).strip().upper() not in ['NA', '', '0'] else 'debit'

    extracted = call_gpt(narration, direction)
    results.append(extracted)
    time.sleep(0.5)  # Avoid rate limiting

# Merge with original
extracted_df = pd.DataFrame(results)
final_df = pd.concat([df, extracted_df], axis=1)
final_df.to_csv("statement_with_extracted_features.csv", index=False)
````

````python
import pandas as pd
import time
import json
from openai import AzureOpenAI

AZURE_OPENAI_KEY = ""
AZURE_OPENAI_ENDPOINT = ""
AZURE_DEPLOYMENT_NAME = ""  # e.g., gpt-4o-mini
AZURE_API_VERSION = ""  # or whatever you're using

# Initialize client
client = AzureOpenAI(
    azure_endpoint=AZURE_OPENAI_ENDPOINT,
    api_key=AZURE_OPENAI_KEY,
    api_version=AZURE_API_VERSION,
)

# Load the data
file_path = r"C:\Users\VIVEK YADAV\Desktop\nextbill\bank_data_for_ai\Zenedl\bank_statement\Bank_statementFY2024-25.xls"
df = pd.read_excel(file_path, header=21, usecols="A:G", nrows=172)
df.columns = ['Date', 'Transaction Details', 'Reference Number', 'Date 2', 'Withdrawal Amt', 'Deposit Amt', 'Balance']
df = df.dropna(how='all')
df['Withdrawal Amt'] = df['Withdrawal Amt'].fillna(0)
df['Deposit Amt'] = df['Deposit Amt'].fillna(0)

# Define GPT function
def call_gpt(narration, direction):
    prompt = f"""
You are a financial data extractor. Given a bank transaction narration, extract the following:

- sender_or_reciever_info: who is the sender or receiver (or UPI ID if available)
- transaction_type: method of transaction like NEFT, UPI, IMPS, etc.
- currency: if foreign currency involved, name it (e.g. USD, EUR)
- exchange_rate: if mentioned, give exchange rate as number

The transaction direction is '{direction}' (either "credit" or "debit"). 
You can determine transaction direction by checking if credit/deposit has value or debit/withdrawl has value

Narration:
\"{narration}\"

Respond only with a valid JSON dictionary like:
{{
  "sender_or_reciever_info": "...",
  "transaction_type": "...",
  "currency": "...",
  "exchange_rate": ...
}}
"""
    try:
        response = client.chat.completions.create(
            model=AZURE_DEPLOYMENT_NAME,
            messages=[{"role": "user", "content": prompt}],
            temperature=0.2,
            max_tokens=200,
        )
        content = response.choices[0].message.content.strip()
        return json.loads(content)
    except Exception as e:
        print(f"Error processing: {narration[:30]} -> {e}")
        return {
            "sender_or_reciever_info": "nan",
            "transaction_type": "nan",
            "currency": "nan",
            "exchange_rate": "nan"
        }

# Process rows and collect outputs
results = []
for i, row in df.iterrows():
    narration = row["Transaction Details"]
    direction = "credit" if row["Deposit Amt"] > 0 else "debit"
    
    result = call_gpt(narration, direction)
    results.append(result)
    
    # Optional: sleep to avoid rate limits
    time.sleep(1)

# Convert extracted features into a DataFrame
extracted_df = pd.DataFrame(results)

# Combine with original data
combined_df = pd.concat([df.reset_index(drop=True), extracted_df], axis=1)

# Save output
combined_df.to_csv("Bank_Statement_with_Extracted_Features.csv", index=False)
print("âœ… Extraction complete. File saved as Bank_Statement_with_Extracted_Features.csv.")
````
