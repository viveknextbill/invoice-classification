```python
import pandas as pd

# Path to the Excel file
file_path = r"C:\Users\VIVEK YADAV\Desktop\nextbill\bank_data_for_ai\Zenedl\bank_statement\Bank_statementFY2024-25.xls"

# Read the Excel file up to row 193 and handle the structure
df = pd.read_excel(file_path, header=21, skiprows=range(0), usecols="A:G", nrows=172)

# Clean the DataFrame
df.columns = ['Date', 'Transaction Details', 'Reference Number', 'Date 2', 'Withdrawal Amt', 'Deposit Amt', 'Balance']  # Rename columns based on structure

# Drop any unnamed columns or extra rows
df = df.dropna(axis=1, how='all')  # Remove columns where all values are NaN

# Clean up any rows that might have partial or corrupt data
df = df.dropna(how='all')  # Remove rows with all NaN values

# Replace NaN in 'Withdrawal Amt' and 'Deposit Amt' columns with 0
df['Withdrawal Amt'] = df['Withdrawal Amt'].fillna(0)
df['Deposit Amt'] = df['Deposit Amt'].fillna(0)

# Display the cleaned DataFrame
df
```
