```python
import camelot
import pandas as pd

pdf_path = r"C:\Users\VIVEK YADAV\Desktop\nextbill\bank_data_for_ai\Amethyst\1. Bank Statement\August\579251377_RAHULBOH_06092024155712.pdf"
#pdf_path = r"C:\Users\VIVEK YADAV\Desktop\nextbill\bank_data_for_ai\Amethyst\1. Bank Statement\June\579251377_RAHULBOH_06072024134133.pdf"

def clean_multiline_cells(df):
    return df.applymap(lambda x: x.replace('\n', ' ') if isinstance(x, str) else x)

def extract_bank_statement_tables(pdf_path, output_csv="output.csv", output_excel="output.xlsx"):
    # Extract tables
    tables = camelot.read_pdf(pdf_path, pages='all', flavor='lattice', strip_text='\n')
    
    if not tables or len(tables) == 0:
        print("No tables found.")
        return

    # Get header from the first table's first row
    header = tables[0].df.iloc[0]

    # Extract data from each table, skipping the header row (row 0)
    data_frames = [table.df.iloc[1:] for table in tables]
    
    # Concatenate all dataframes
    full_data = pd.concat(data_frames, ignore_index=True)

    # Clean up cell content (remove newlines)
    full_data = clean_multiline_cells(full_data)

    # Assign proper headers
    full_data.columns = header

    # Save outputs
    full_data.to_csv(output_csv, index=False)
    full_data.to_excel(output_excel, index=False)
    print(f"Saved to {output_csv} and {output_excel}")

extract_bank_statement_tables(pdf_path)
```
